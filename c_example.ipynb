```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
import torch

from tcnAutoencoder import TCNAutoencoder
```

```python
tensor1 = np.random.randint(0, 11, (1800, 3))
tensor2 = np.random.randint(0, 11, (1800, 3))
tensor3 = np.random.randint(0, 11, (1800, 3))
tensor4 = np.random.randint(0, 11, (1800, 3))
tensor5 = np.random.randint(0, 11, (1800, 3))

abnormal_tensor = np.random.randint(20, 31, (1800, 3))
tensors = np.array([tensor1, tensor2, tensor3, tensor4, tensor5, abnormal_tensor])
flattened_tensors = np.array([tensor.flatten() for tensor in tensors])
e_distance_matrix = squareform(pdist(flattened_tensors, metric='euclidean'))
linkage_matrix = linkage(flattened_tensors, method='complete')
plt.figure(figsize=(10, 7))
dendro = dendrogram(linkage_matrix, labels=[f'Tensor {i+1}' for i in range(len(tensors))])
plt.title('Sample Dendrogram for Hierarchical Clustering')
plt.xlabel('Tensor index')
plt.ylabel('Distance')
plt.show()
```
